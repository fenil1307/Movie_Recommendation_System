```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('./Datasets/tmdb_5000_movies.csv')
credits = pd.read_csv('./Datasets/tmdb_5000_credits.csv') 
```

```python
movies
```

```python
credits
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies
```

```python
movies.isnull().sum()
```

```python
movies.dropna()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L
```

```python
movies['genres']
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords']
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies['keywords']
```

```python
movies['cast']
```

```python
movies['cast'][0]
```

```python
movies['cast'] = movies['cast'].apply(convert)
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
movies['cast'][0]
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head(3)
```

```python
movies['overview'].dtypes
```

```python
def split_text(x):
    if isinstance(x, str):
        return x.split()
    else:
        return []
```

```python
movies['overview'] = movies['overview'].apply(split_text)
```

```python
movies['overview'] 
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
len(movies['tags'][0])
```

```python
len(movies['tags'][1])
```

```python
movies_Final = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
movies_Final['tags'] = movies_Final['tags'].apply(lambda x: " ".join(x))
movies_Final.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
pas = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(pas.stem(i))
    return " ".join(y)
```

```python
movies_Final['tags'] = movies_Final['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(movies_Final['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = movies_Final[movies_Final['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(movies_Final.iloc[i[0]].title)
```

```python
recommend('Spectre')
```
